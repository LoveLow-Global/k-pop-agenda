# Note

This code is intended to check the missing articles, after we have obtained the text data from the new articles.

Articles are usually missing because it has been deleted from the website and therefore unable to find anymore.

Before running the code, change `base_dir` to your folder location where the new articles are at

```python
import os
import datetime
```

```python
def find_missing_articles(base_dir):
    """
    Finds missing article files.

    Args:
        base_dir: The base directory containing the article files.
    """
    start_date = datetime.date(2024, 1, 1)
    end_date = datetime.date(2024, 12, 31)
    max_articles_per_day = 10

    missing_articles = []
    current_date = start_date
    while current_date <= end_date:
        date_str = current_date.strftime("%Y%m%d")
        for article_num in range(1, max_articles_per_day + 1):
            file_path = os.path.join(base_dir, f"{date_str}_article_{article_num}.txt")
            if not os.path.exists(file_path):
                missing_articles.append(file_path)
        current_date += datetime.timedelta(days=1)

    return missing_articles


if __name__ == "__main__":
    base_dir = "C:/Users/WINDOWS 11/Desktop/kpop_agenda/Step1/Articles"
    missing = find_missing_articles(base_dir)
    if missing:
        print("Missing article files:")
        for file_path in missing:
            print(file_path)
    else:
        print("No missing article files found.")
```
